```python
#noi com dien
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

time=ctrl.Antecedent(np.arange(10,120,1),'time')
rice = ctrl.Antecedent(np.arange(100,1000,1),'rice')
power=ctrl.Consequent(np.arange(0,101,1),'power')

time['rat lau']=fuzz.trimf(time.universe,[10,15,20])
time['lau']=fuzz.trimf(time.universe,[15,20, 40])
time['vua']=fuzz.trimf(time.universe,[20,40,60])
time['nhanh']=fuzz.trimf(time.universe,[40,60,90])
time['rat nhanh']=fuzz.trimf(time.universe,[60,90,120])

rice['rat it']=fuzz.trimf(rice.universe,[0,150, 300])
rice['it']=fuzz.trimf(rice.universe,[150, 300,500])
rice['vua']=fuzz.trimf(rice.universe,[ 300,500,750])
rice['nhieu']=fuzz.trimf(rice.universe,[500,750,900])
rice['rat nhieu']=fuzz.trimf(rice.universe,[750,900,1000])

power['rat nho'] = fuzz.trimf(power.universe,[0,0,20])
power['nho'] = fuzz.trimf(power.universe,[0,20, 40])
power['trung binh'] = fuzz.trimf(power.universe,[ 20,40,60])
power['lon'] = fuzz.trimf(power.universe,[40,60,90])
power['rat lon'] = fuzz.trimf(power.universe,[60,90,101])

rule1=ctrl.Rule(time['rat lau']&rice['rat it'],power['rat nho'])
rule2=ctrl.Rule(time['rat lau']&rice['it'],power['rat nho'])
rule3=ctrl.Rule(time['rat lau']&rice['vua'],power['nho'])
rule4 =ctrl.Rule(time['rat lau']&rice['nhieu'], power['trung binh'])
rule5=ctrl.Rule(time['rat lau']&rice['rat nhieu'],power['trung binh'])

rule6=ctrl.Rule(time['lau']&rice['rat it'],power['rat nho'])
rule7=ctrl.Rule(time['lau']&rice['it'],power['nho'])
rule8=ctrl.Rule(time['lau']&rice['vua'],power['trung binh'])
rule9=ctrl.Rule(time['lau']&rice['nhieu'],power['trung binh'])
rule10=ctrl.Rule(time['lau']&rice['rat nhieu'],power['lon'])

rule11=ctrl.Rule(time['vua']&rice['rat it'],power['nho'])
rule12=ctrl.Rule(time['vua']&rice['it'],power['nho'])
rule13=ctrl.Rule(time['vua']&rice['vua'],power['trung binh'])
rule14=ctrl.Rule(time['vua']&rice['nhieu'],power['trung binh'])
rule15=ctrl.Rule(time['vua']&rice['rat nhieu'],power['lon'])

rule16=ctrl.Rule(time['nhanh']&rice['rat it'],power['trung binh'])
rule17=ctrl.Rule(time['nhanh']&rice['it'],power['trung binh'])
rule18=ctrl.Rule(time['nhanh']&rice['vua'],power['lon'])
rule19=ctrl.Rule(time['nhanh']&rice['nhieu'],power['lon'])
rule20=ctrl.Rule(time['nhanh']&rice['rat nhieu'],power['rat lon'])

rule21=ctrl.Rule(time['rat nhanh']&rice['rat it'],power['trung binh'])
rule22=ctrl.Rule(time['rat nhanh']&rice['it'],power['lon'])
rule23=ctrl.Rule(time['rat nhanh']&rice['vua'],power['lon'])
rule24=ctrl.Rule(time['rat nhanh']&rice['nhieu'],power['rat lon'])
rule25=ctrl.Rule(time['rat nhanh']&rice['rat nhieu'],power['rat lon'])


powering_ctrl = ctrl.ControlSystem([rule1,rule2,rule3,rule4,rule5,rule6,rule7,rule8,rule9,rule10,rule11,rule12,rule13,rule14,rule15,rule16,rule17,rule18,rule19,rule20,rule21,rule22,rule23,rule24,rule25])
powering = ctrl.ControlSystemSimulation(powering_ctrl)

powering.input['rice']= 800
powering.input['time']= 90
powering.compute()
print(powering.output['power'])
power.view()
```
